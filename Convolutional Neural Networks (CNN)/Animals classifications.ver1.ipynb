```python
import os 
import matplotlib.pyplot as plt 
import seaborn as sns
import cv2
import numpy as np 
```

```python
labels = ['dog','horse','elephant','butterfly','chicken','cat','cow','sheep','spider','squirrel']
img_size = 224
def find_datafunc(data_dir):
    data = []
    for label in labels: 
        path = os.path.join(data_dir,label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr= cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
                cvtColor = cv2.cvtColor(img_arr,cv2.COLOR_BGR2RGB)
                resize_arr = cv2.resize(cvtColor,(224,224))
                data.append([resize_arr,class_num])
            except Exception as e: 
                print(e)        
    return np.array(data)
```

```python
data = find_datafunc('C:/Users/Lee Jong Ann/Desktop/animals/raw-img/')
```

```python
l = []
for i in data: 
    l.append(labels[i[1]])
fig,ax = plt.subplots()
fig.set_size_inches(17,7)
plt.sca(ax)
plt.xticks(rotation=30)
sns.set_style('dark')
sns.countplot(l)
```
