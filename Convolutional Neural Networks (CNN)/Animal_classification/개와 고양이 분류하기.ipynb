```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import rlcompleter, readline
readline.parse_and_bind('tab:complete')
```

```python
filenames = os.listdir('C:/Users/Lee Jong Ann/Desktop/dataset/CatsandDogs/dataset/training_set')
```

```python
filelist = []

for filename in filenames: 
    animal_class = filename.split('.')[0]
    if animal_class == 'dog': 
        filelist.append(1)
    else: 
        filelist.append(0)
                            
        
```

```python
animal_frame = pd.DataFrame({ 


    'filename':filenames,
    'category':filelist
})
```

```python
sns.factorplot(data=animal_frame,kind='count',x='category')
```
